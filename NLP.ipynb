```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, accuracy_score
import pandas as pd
import nltk
```

```python
import datasets as dataset

dataset_instance=dataset.load_dataset("winvoker/turkish-sentiment-analysis-dataset")
train_df=pd.DataFrame(dataset_instance["train"])
test_df=pd.DataFrame(dataset_instance["test"])
train_df.head()
test_df.head()
```

```python
x_train=train_df["text"]
y_train=train_df["label"]
x_test=test_df["text"]
y_test=test_df["label"]
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
turkish_stop_words=stopwords.words("turkish")
```

```python
vectorizer=TfidfVectorizer(max_features=5000, stop_words=turkish_stop_words)
x_train_tfidf=vectorizer.fit_transform(x_train)
x_test_tfidf=vectorizer.transform(x_test)
```

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=200),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "Support Vector Machine": SVC(),
    "Naive Bayes": MultinomialNB()
}
```

```python
for model_name, model in models.items():
    print(f"Logistik Regression: {LogisticRegression}")
    print(f"Random Forest: {RandomForestClassifier}")
    print(f"Gradient Boosting: {GradientBoostingClassifier}")
    print(f"Support Vector Machine: {SVC}")
    print(f"Naive Bayes: {MultinomialNB}")
    model.fit(x_train_tfidf, y_train)
    y_pred = model.predict(x_test_tfidf)
    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred)
    print(f"Doğruluk Oranı: {accuracy:.2%}")
    print("Classification Report:")
    print(report)
```
